## 5. Describir variables numéricas y categóricas
Usar describe() para variables numéricas y value_counts() para variables categóricas.

```python
# Numerical variables
print("RESUMEN VARIABLES NUMÉRICAS")
display(df.describe())

# Categorical values (Example for a columnn)
# df['category_column'].value_counts()

print("\nVARIABLES CATEGÓRICAS")

cat_cols = df.select_dtypes(include=["object", "str"]).columns

for col in cat_cols:
    print(f"\n--- Conteo de valores para: {col} ---")
    display(df[[col]].value_counts().to_frame(name='Frecuencia'))
```

## 7. Análisis univariado de variables numéricas
Realizar histogramas y boxplots para analizar la distribución de las variables numéricas.

```python
variables_preguntas = ['grades', 'gaming_hours', 'addiction_score', 'study_hours', 'sleep_hours']

print("=========================================")
print("  GENERANDO HISTOGRAMAS DE LAS PREGUNTAS ")

for col in variables_preguntas:
    plt.figure(figsize=(6, 4))
    # Usamos un color azul universitario muy limpio con la curva KDE activada
    sns.histplot(data=df, x=col, kde=True, color='#2b5c8f')
    
    # Formateamos los títulos para que reemplacen los guiones bajos por espacios elegantes
    titulo_limpio = col.replace("_", " ").title()
    plt.title(f'Distribución Global de {titulo_limpio}', fontsize=12, pad=10)
    plt.xlabel(titulo_limpio, fontsize=10)
    plt.ylabel('Frecuencia (Estudiantes)', fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    
    plt.show()

# =====================================================================
# 2. BOXPLOTS: ¿Hay valores atípicos (outliers) o asimetrías extremas?
# =====================================================================
print("\n========================================")
print("   GENERANDO BOXPLOTS DE LAS PREGUNTAS  ")
print("========================================")

for col in variables_preguntas:
    plt.figure(figsize=(6, 3))
    # Usamos un color verde sutil para diferenciar visualmente que cambiamos de tipo de gráfico
    sns.boxplot(data=df, x=col, color='#4f9a74')
    
    titulo_limpio = col.replace("_", " ").title()
    plt.title(f'Diagrama de Caja y Outliers de {titulo_limpio}', fontsize=12, pad=10)
    plt.xlabel(titulo_limpio, fontsize=10)
    plt.grid(axis='x', linestyle='--', alpha=0.5)
    
    plt.show()
```

## 8. Análisis univariado de variables categóricas
Realizar gráficos de barras para analizar la frecuencia de las variables categóricas.

```python
variables_cat_preguntas = ['gaming_genre', 'stress_level', 'gender']

print("========================================")
print("  GENERANDO GRÁFICOS DE BARRAS (CAT)   ")
print("========================================")

for col in variables_cat_preguntas:
    plt.figure(figsize=(7, 4))
    
    # Ordenamos las barras de mayor a menor frecuencia automáticamente para que sea más legible
    orden_barras = df[col].value_counts().index
    
    # Añadimos hue=col y legend=False para cumplir con las nuevas versiones de Seaborn y evitar el Warning
    ax = sns.countplot(data=df, x=col, order=orden_barras, hue=col, palette='viridis', legend=False)
    
    # Añadimos el número exacto de estudiantes encima de cada barra para dar más rigor estadístico
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', 
                    xytext=(0, 5), 
                    textcoords='offset points', 
                    fontsize=10, fontweight='bold')

    # Formateamos títulos y etiquetas limpiando los guiones bajos
    titulo_limpio = col.replace("_", " ").title()
    plt.title(f'Frecuencia de Estudiantes por {titulo_limpio}', fontsize=13, pad=15)
    plt.xlabel(titulo_limpio, fontsize=11)
    plt.ylabel('Cantidad de Estudiantes', fontsize=11)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    
    plt.show()
```

## 9. Análisis bivariado entre variables
Explorar relaciones entre variables numéricas y categóricas mediante scatterplots y tablas cruzadas.

```python


# %%
print("=========================================================================")
# PREGUNTA 1: ¿Más horas de gaming se asocian con peores notas?
print(" RESOLVIENDO PREGUNTA 1: GAMING HOURS VS GRADES (Numérica vs Numérica) ")
print("=========================================================================")

plt.figure(figsize=(7, 4.5))
# regplot dibuja los puntos y calcula la recta de regresión lineal automáticamente
sns.regplot(data=df, x='gaming_hours', y='grades', 
            scatter_kws={'alpha':0.3, 'color': '#2b5c8f'}, 
            line_kws={'color': 'red', 'linewidth': 2})

plt.title('Impacto de las Horas de Gaming Semanales en las Calificaciones', fontsize=12, pad=12)
plt.xlabel('Horas de Gaming (Semanales)', fontsize=10)
plt.ylabel('Calificaciones (Grades)', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()


print("\n=========================================================================")
# PREGUNTA 2: ¿Qué género de videojuego se asocia con mayor o menor rendimiento?
print("  RESOLVIENDO PREGUNTA 2: GAMING GENRE VS GRADES (Categórica vs Numérica) ")
print("=========================================================================")

plt.figure(figsize=(8, 4.5))
# Ordenamos los géneros por su mediana de nota para que el gráfico sea súper limpio
orden_generos = df.groupby('gaming_genre')['grades'].median().sort_values(ascending=False).index

sns.boxplot(data=df, x='gaming_genre', y='grades', order=orden_generos, hue='gaming_genre', palette='Set2', legend=False)

plt.title('Rendimiento Académico (Grades) según el Género de Videojuego Preferido', fontsize=12, pad=12)
plt.xlabel('Género de Videojuego', fontsize=10)
plt.ylabel('Calificaciones (Grades)', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()


print("\n=========================================================================")
# PREGUNTA 3: ¿El nivel de adicción modera la relación entre gaming y notas?
print("     RESOLVIENDO PREGUNTA 3: EL EFECTO MODERADOR DE LA ADICCIÓN          ")
print("=========================================================================")

plt.figure(figsize=(9, 5))
# Usamos un gráfico de dispersión donde el color de cada punto representa el "Addiction Score"
scatter = plt.scatter(data=df, x='gaming_hours', y='grades', c='addiction_score', 
                      cmap='viridis', alpha=0.5, edgecolors='none')

# Añadimos una barra lateral de color para entender la escala de adicción
cbar = plt.colorbar(scatter)
cbar.set_label('Puntuación de Adicción (Addiction Score)', fontsize=10, labelpad=10)

plt.title('Interacción entre Horas de Juego, Adicción y Calificaciones Finales', fontsize=12, pad=12)
plt.xlabel('Horas de Gaming (Semanales)', fontsize=10)
plt.ylabel('Calificaciones (Grades)', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()


print("\n=========================================================================")
# TABLAS CRUZADAS: Relación entre variables cualitativas
print("      ANÁLISIS ADICIONAL: TABLA CRUZADA GÉNERO DE JUEGO VS ESTRÉS        ")
print("=========================================================================")

# Cruzamos el tipo de juego con el nivel de estrés percibido para buscar patrones conductuales
tabla_cruzada = pd.crosstab(
    df['gaming_genre'],
    df['stress_level'],
    margins=True, # Añade los totales de filas y columnas automáticamente
    margins_name="Total"
)

display(tabla_cruzada)
```

## 10. Visualización de correlaciones
Calcular y visualizar la matriz de correlación entre variables numéricas usando un heatmap.

```python
print("=========================================================================")
# FILTRADO DE VARIABLES PARA EVITAR RUIDO
# Excluimos 'student_id' ya que no tiene significado matemático para la correlación
print(" 1. GENERANDO MATRIZ DE CORRELACIÓN GENERAL (Limpia sin IDs) ")
print("=========================================================================")

variables_numericas = df.select_dtypes(include="number").drop(columns=['student_id'], errors='ignore')
corr_general = variables_numericas.corr()

plt.figure(figsize=(10, 7))
sns.heatmap(
    corr_general,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5,
    annot_kws={"size": 9}
)
plt.title('Matriz de Correlación de Variables Numéricas', fontsize=13, pad=15)
plt.show()


print("\n=========================================================================")
# PREGUNTA 4: ¿Cuáles son los factores más correlacionados con las calificaciones?
print(" RESOLVIENDO PREGUNTA 4: RANKING DE CORRELACIÓN RESPECTO A LAS NOTAS (Grades) ")
print("=========================================================================")

# Extraemos únicamente la columna de correlaciones de 'grades' y quitamos su autocorrelación (grades vs grades)
corr_notas = corr_general[['grades']].drop(index=['grades']).sort_values(by='grades', ascending=False)

plt.figure(figsize=(4, 6))

# Usamos una paleta divergente de Rojo a Verde (RdYlGn) para identificar rápido lo positivo y negativo
sns.heatmap(
    corr_notas, 
    annot=True, 
    cmap='RdYlGn', 
    fmt='.2f', 
    cbar=False, 
    linewidths=0.5, 
    annot_kws={"size": 12, "weight": "bold"}
)

plt.title('Factores más Correlacionados\ncon las Calificaciones (Grades)', fontsize=12, pad=15)
plt.ylabel('Variables del Dataset', fontsize=10)
plt.show()
```

## Las Preguntas de este análisis son:






¿Más horas de gaming se asocian con peores notas?
### Respuesta 1:
Sí, existe una relación negativa clara entre el tiempo dedicado a los videojuegos y el rendimiento académico. A medida que incrementan las horas semanales de juego, las calificaciones tienden a disminuir notablemente, acentuándose el descenso a partir de las 15 horas de gaming.

```python
plt.figure(figsize=(8, 5))
# Usamos regplot para ver los puntos junto con la línea de tendencia estimada
sns.regplot(data=df, x='gaming_hours', y='grades', 
            scatter_kws={'alpha':0.4, 'color': '#2b5c8f'}, 
            line_kws={'color': 'red', 'linewidth': 2})

plt.title('Relación entre Horas de Gaming Semanales y Calificaciones', fontsize=14, pad=15)
plt.xlabel('Horas de Gaming (Semanales)', fontsize=12)
plt.ylabel('Calificaciones (Grades)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

¿Qué género de videojuego se asocia con mayor o menor rendimiento?
### Respuesta 2:
El rendimiento académico varía según las preferencias del tipo de juego. Los estudiantes orientados a géneros competitivos o de alta demanda de tiempo continuo, como los **FPS (First-Person Shooters)** y **Strategy**, registran los promedios de calificaciones más bajos. En contraste, aquellos que se decantan por videojuegos **Casual** o **RPG** logran mantener un rendimiento significativamente más alto.

```python
plt.figure(figsize=(9, 5))
# Ordenamos las cajas de mayor a menor rendimiento promedio para mejorar la visualización
orden_generos = df.groupby('gaming_genre')['grades'].median().sort_values(ascending=False).index

sns.boxplot(data=df, x='gaming_genre', y='grades', order=orden_generos, palette='Set2')

plt.title('Distribución de Calificaciones por Género de Videojuego Preferido', fontsize=14, pad=15)
plt.xlabel('Género de Videojuego', fontsize=12)
plt.ylabel('Calificaciones (Grades)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

¿El nivel de adicción modera la relación entre gaming y notas?
### Respuesta 3:
Efectivamente, el nivel de adicción actúa como un factor crítico. Al mapear el nivel de adicción (`addiction_score`) mediante una escala de color sobre la relación gaming-notas, observamos que los estudiantes que sufren el desplome académico más severo no solo juegan más horas, sino que concentran los índices de adicción más elevados. Una puntuación alta de adicción agrava el impacto negativo del gaming sobre las calificaciones.

```python
plt.figure(figsize=(9, 6))
# Usamos un scatterplot donde el color (hue) representa dinámicamente el nivel de adición
scatter = plt.scatter(data=df, x='gaming_hours', y='grades', c='addiction_score', 
                      cmap='viridis', alpha=0.6, edgecolors='none')

cbar = plt.colorbar(scatter)
cbar.set_label('Puntuación de Adicción (Addiction Score)', fontsize=11, labelpad=10)

plt.title('Moderación de la Adicción en la Relación Gaming vs Calificaciones', fontsize=14, pad=15)
plt.xlabel('Horas de Gaming (Semanales)', fontsize=12)
plt.ylabel('Calificaciones (Grades)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

¿Cuáles son los factores más correlacionados con las calificaciones? 
### Respuesta 4:
Analizando las correlaciones lineales con la variable objetivo `grades`, los resultados demuestran que:
1. **Factores Positivos:** Las horas de estudio (`study_hours`) y el porcentaje de asistencia (`attendance`) son los motores principales del éxito académico.
